```python
# import tensorflow as tf
# from keras.layers import Dense,LSTM
# from keras.models import Sequential
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.metrics import f1_score,mean_squared_error,r2_score,root_mean_squared_error,precision_score
import datetime as dt
import numpy as np
import pickle 
import matplotlib.pyplot as plt
```

```python
def gbm(mu,n,T,M,S0,sigma):
    # calc each time step
    dt = T/n
    
    # simulation using numpy arrays
    St = np.exp(
        (mu - sigma ** 2 / 2) * dt
        + sigma * np.random.normal(0, np.sqrt(dt), size=(M,n)).T
    )
    
    # include array of 1's
    St = np.vstack([np.ones(M), St])
    
    # multiply through by S0 and return the cumulative product of elements along a given simulation path (axis=0).
    St = S0 * St.cumprod(axis=0)
    
    
    
    # Define time interval correctly
    time = np.linspace(0,T,n+1)
    
    # Require numpy array that is the same shape as St
    tt = np.full(shape=(M,n+1), fill_value=time).T

    return tt,St
```

```python
mu = 0.17
# number of steps
n = 1000
# time in years
T = 1/252
# number of sims
M = 100
# initial stock price
S0 = 22614
# volatility
sigma = 0.2

tt,St = gbm(mu,n,T,M,S0,sigma)
```

```python
open_prices = [22614,22900,22800]
def data_creator_using_gbm(open_prices):
    data = pd.DataFrame(columns=['open_price','lower_bound_gbm','upper_bound_gbm'])
    min_array_mean = np.array([])
    max_array_mean = np.array([])
    for x in open_prices:
        tt,St = gbm(mu,n,T,M,x,sigma)
        min_array = np.array([])
        max_array = np.array([])
        for y in range(100):
            array1 = []
            for x in range(1001):
                array1.append(St[x][y])
            min = np.array(array1).min()
            max = np.array(array1).max()
            min_array = np.append(min_array,min)
            max_array = np.append(max_array,max)
        min_array_mean=np.append(min_array_mean,min_array.mean())
        max_array_mean=np.append(max_array_mean,max_array.mean())
    data['open_price'] = open_prices
    data['lower_bound_gbm'] = min_array_mean
    data['upper_bound_gbm'] = max_array_mean
    return data 
data_creator_using_gbm(open_prices)
```
